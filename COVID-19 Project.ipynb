### Q.3) In which Region, maximum number of Confirmed cases were recorded ?

```python
#df.groupby('Region').Confirmed.sum().sort_values(ascending = False).head(20)
```

```python
data.head(2)
```

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending = False).head(20)
```

```python
# Extracting the top 20 regions and their confirmed cases
top_regions = data.groupby('Region').Confirmed.sum().sort_values(ascending=False).head(20)

# Creating a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(top_regions.index, top_regions.values, color='skyblue')
plt.xlabel('Region')
plt.ylabel('Confirmed Cases')
plt.title('Top 20 Regions with Confirmed Cases')
plt.xticks(rotation=90)

# Adding data labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black', fontsize=6)

plt.tight_layout()
plt.show()
```
